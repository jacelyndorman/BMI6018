```python
#Use the below datasets create any four visualization summarizing the datasets. 
#You can use either Matplotlib, Seaborn or any other python based visualization package. 
#Be creative!. Here are some example visualizations: https://blog.hubspot.com/marketing/types-of-graphs-for-data-visualization. 
#Remember to label and have legends for your graphs. Add comments to describe what you are trying to graph

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = pd.read_csv("diabetic_data_edited.csv")
```

```python
#Visualization #1: Line Graph

sns.lineplot(data = data , x = "age", y = "num_medications", hue = "gender").set(title = "Average Number of Medications by Age")
plt.xlabel("Age Group")
plt.ylabel("Number of Medications")

plt.plot() 


# This graph is showing the average number of medications for diabetes patients by age group.
# As you can see, the number of medications increases until about age 60-70 and then the average number decreases.
# The averages are fairly similar between males and females, although males on average had slightly fewer medications than females. 
# There is more variability in the number of medications amongst the younger ages, as shown by the bands around the lines. This could also be due to a smaller dataset for this age group.
# Note: There were 3 data points with an "Unknown/Invalid" gender. These data points were not very helpful as their own category, so I removed them.

```

```python
#Visualization #2: Bar Chart

sns.set(font_scale = 0.75) #Loweing the font size because some of the x axis labels were overlapping
sns.barplot(data = data, x = "race", y = "num_procedures", hue = "gender")
plt.title("Number of Procedures by Race")
plt.xlabel("Race")
plt.ylabel("Number of Procedures")
plt.show()

# This graph shows the number of procedures for diabetes patients by race (grouped by genders). 
# It looks like Hispanic and African American patients have fewer procedures on average than Caucasian patients. 
# One could hypothesize that Hispanics and African Americans do not have the same access to procedures that Caucasians do.
# Interestingly, males seem to have more procedures done than females, regardless of race. 
# I wonder how this relates to the previous data showing that males on average take fewer medications for diabetes than females.
```

```python
#Visualization #3: Heatmap

pivoted = data.pivot_table(index = "age", columns = "time_in_hospital", values = "num_medications")
ax = sns.heatmap(pivoted, annot = True, linewidth = 0.5)
ax.set(xlabel = "Time In Hospital (Days)", ylabel = "Age")
ax.set_title("Number of Medications per Hospitalization Time")
plt.show()

#This visual is intended to show a correlation between the time a diabetes patient spends in the hospital and the number of medications they take.
#The lighter color indicates a higher number of medications, and the darker color indicates a lower number of medications. 
#As you can see, number of medications is correlated with time spent in hospital. The more medications a patient takes, the more time they spend in the hospital.
#I am not sure if this is the number of medications they took before coming to the hospital or after coming to the hospital.
#However, it makes sense to me either way. If someone is already on many medications, they are likely not healthy and will need to spend more time in the hospital.
#On the other hand, if someone spends a lot of time in the hospital, their condition is likely very serious and they will need more medications to recover.
```

```python
#Visualization #4: Pie Chart

sns.displot(data = data, x = "num_medications")
plt.title("Distribution of Medications")
plt.show()

#This visual shows the distribution of number of medications for diabetes patients. The count shows how many people have that number of medications. 
#You can see that most people have between 10 and 20 medications. Very few have over 40 medications. 

```
